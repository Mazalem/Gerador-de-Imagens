### Importação de bibliotecas

```python
import time                         # Para medir tempo de execução, se necessário
import keras_cv                     # Biblioteca que contém modelos pré-treinados e ferramentas de visão computacional
from tensorflow import keras        # API de alto nível para redes neurais
import matplotlib.pyplot as plt     # Biblioteca para visualização e gráficos
```

### Criação do modelo Stable Diffusion

```python
# Usa precisão mista (float16 + float32) para acelerar e economizar memória
keras.mixed_precision.set_global_policy("mixed_float16")
```

```python
# Cria uma instância do modelo Stable Diffusion configurada para imagens 512x512
modelo = keras_cv.models.StableDiffusion(img_width=512, img_height=512)
```

### Função para exibir imagens

```python
def plot_images(imagens):
    plt.figure(figsize=(20,20))                 # Define o tamanho da figura
    for i in range(len(imagens)):               # Itera sobre cada imagem gerada
        ax = plt.subplot(1, len(imagens), i+1)   # Cria subplots lado a lado
        plt.imshow(imagens[i])                   # Mostra a imagem no subplot
        plt.axis("off")                          # Remove os eixos para visualização limpa
```

### Teste

```python
imagens = modelo.text_to_image(
    "INSIRA SEU PROMPT AQUI (EM INGLÊS)",
    batch_size=3
)
```

```python
plot_images(imagens)
```
